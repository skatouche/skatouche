```r
if (!require(ggplot2)) install.packages("ggplot2")
library(ggplot2)

# Création d'un jeu de données exemple
set.seed(123)
data <- data.frame(
  category = rep(c('A', 'B', 'C'), each = 100),
  x = rep(1:100, times = 3),
  y = rep(1:100, times = 3) + rnorm(300, sd = 20)
)

# Création du graphique complexe
p <- ggplot(data, aes(x = x, y = y, color = category)) +
  geom_point() +                        # Ajout des points
  geom_smooth(method = "lm") +          # Ajout d'une ligne de tendance
  facet_wrap(~ category) +              # Facettage par catégorie
  theme_minimal() +                     # Application d'un thème minimaliste
  labs(title = "Graphique Complexe avec ggplot2",
       x = "Axe X",
       y = "Axe Y",
       color = "Catégorie")             # Ajout de titres et labels

# Affichage du graphique
print(p)
```
